```python
def getResponseTime(data):
    """Computes the response times for every answer except the first and last of the day"""  
    data.sort_values(by="SubmitDateTime")
    all_dates_times = [datetime.datetime.strptime(i, '%Y-%m-%d %H:%M:%S.%f') for i in data["SubmitDateTime"]]
    unique_dates = set([datetime.datetime.strftime(d, '%Y-%m-%d') for d in all_dates_times])
    unique_dates = sorted(unique_dates)
    RT = []
    D = []
    D.append(0)
    for u in range(1, len(unique_dates)):
        D.append((datetime.datetime.strptime(list(unique_dates)[u], '%Y-%m-%d') - datetime.datetime.strptime(list(unique_dates)[u - 1], '%Y-%m-%d')).days)
        
    for u in unique_dates:
        daily_data = data.loc[data["SubmitDateTime"].str.contains(u)]
        times = [datetime.datetime.strftime(i, '%H:%M:%S.%f') for i in [datetime.datetime.strptime(dd, '%Y-%m-%d %H:%M:%S.%f') for dd in daily_data["SubmitDateTime"]]]
        RT.append([None])
        responseTimes = [str(datetime.datetime.strptime(times[i], '%H:%M:%S.%f') - datetime.datetime.strptime(times[i - 1], '%H:%M:%S.%f')) for i in range(1, len(times))]
        RT.append(responseTimes)
        print(responseTimes)
    
    NRT = list(itertools.chain.from_iterable(RT))
    newdata = data.assign(ExTime = NRT)
    return NRT, D, newdata
```
